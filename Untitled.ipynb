```python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import skimage as ski
import ipympl
import numpy as np


%matplotlib widget

# load an image
image = iio.imread(uri='colonies-01.tif')

# rotate it by 45 degrees
rotated = ski.transform.rotate(image=image, angle=45)

# display the original image and its rotated version side by side
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(rotated)
```

```python
"""Python script to read an image, resize it, and save it under a different name."""

import imageio.v3 as iio
import ipympl
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski

# read in image
chair = iio.imread(uri="chair.jpg")

# resize the image
new_shape = (chair.shape[0] // 1000, chair.shape[1] // 1000, chair.shape[2])
resized_chair = ski.transform.resize(image=chair, output_shape=new_shape)
resized_chair = ski.util.img_as_ubyte(resized_chair)

# write out image
iio.imwrite(uri="resized_chair.jpg", image=resized_chair)

# display images
fig, ax = plt.subplots()
ax.imshow(chair)
fig, ax = plt.subplots()
ax.imshow(resized_chair)
```
