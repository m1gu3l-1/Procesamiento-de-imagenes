```python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen original
image_path = "maize-root-cluster.jpg"
image = iio.imread(uri=image_path)

# Definir los lmites del recorte (ajustar segn sea necesario)
# Suponemos que sabemos los lmites aproximados de la planta y sus races
# Estos lmites son un ejemplo y deben ser ajustados a la imagen real
start_row, end_row = 50, 400  # Fila de inicio y fin
start_col, end_col = 100, 500  # Columna de inicio y fin

# Recortar la subimagen
subimage = image[start_row:end_row, start_col:end_col]

# Mostrar la subimagen
fig, ax = plt.subplots()
ax.imshow(subimage)
plt.show()



```
